```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Replace 'your_csv_file.csv' with the path to your CSV file

file_path = "../Serverful/results/15,10,10,10==>[2024-04-06]23:16:14.315/result.csv"

dimensions_str = file_path.split('/')[-2]
dimensions_str = dimensions_str.split("==>[")[0]
dimensions = dimensions_str.split(",")  # Split by comma
show_dim = [int(dim) for dim in dimensions]

df3 = pd.read_csv(file_path)

# Calculate the average of execution time and network delay
severful_avg_exe_time = df3['execution_time'].mean()
severful_avg_latency = df3['network_delay'].mean()
severful_avg_response = severful_avg_exe_time + severful_avg_latency

df4 = pd.read_csv('../Serverless/log/S04_15,10,10,10_lambda_test_log_2024-04-06_18-27-43.csv')
severless_avg_exe_time = df4['execution_time'].mean()
severless_avg_latency = df4['network_delay'].mean()
severless_avg_response = severless_avg_exe_time + severless_avg_latency

# Data
categories = ['Serverful', 'Serverless']
response_time = [severful_avg_response, severless_avg_response]
execution_time = [severful_avg_exe_time, severless_avg_exe_time]
latency = [severful_avg_latency, severless_avg_latency]

# Plotting
bar_width = 0.25
index = range(len(categories))

plt.bar(index, execution_time, bar_width, label='Execution Time')
plt.bar([i + bar_width for i in index], latency, bar_width, label='Latency')

plt.plot([i + bar_width / 2 for i in index], response_time, marker='o', linestyle='-', color='r', label='Response Time')

plt.xlabel('Category')
plt.ylabel('Time(s)')
plt.title(f"Severful / Severless Compare Dimension {show_dim[0]} Iteration {show_dim[1]} Qty_loop {show_dim[2]}")
plt.xticks([i + bar_width for i in index], categories)
plt.legend()
plt.show()
```
